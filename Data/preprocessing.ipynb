```python
import pandas as pd
import numpy as np
import os

import warnings
warnings.filterwarnings('ignore')
```

Importación de Datasets

```python
# Definición de ruta de archivos

file_path = './data/raw_data/'
```

Carga de archivo en Excel

```python
def loadData():
    # Get the list of files in the folder
    files_in_folder = os.listdir(file_path)

    # Find the Excel file in the folder
    excel_file = None
    for file in files_in_folder:
        if file.lower().endswith('.xlsx'):
            excel_file = os.path.join(file_path, file)
            break

    if excel_file:
        # Load the Excel file into an ExcelFile object
        xls = pd.ExcelFile(excel_file)

        # Initialize a dictionary to store the DataFrames
        dataframes = {}

        # Iterate through each sheet in the Excel file
        for sheet in xls.sheet_names:
            # Read each sheet into a DataFrame
            df = xls.parse(sheet)
            # Store the DataFrame in the dictionary using the sheet name as the key
            dataframes[sheet] = df

        return dataframes
    else:
        print(f"No Excel file found in the folder: {file_path}")
        return None

# Example usage:
# Suppose you have an Excel file in the folder "my_folder," and you don't know the file name.
# You can call the function like this:
# dataframes = load_excel_and_extract_dataframes_in_folder("my_folder")
# dataframes will contain a dictionary with each DataFrame associated with a sheet in the Excel file.
```
