```python
cols = [['Nº pers.', 'CP Vivienda', 'CP Trabajo', 'Ubicación', 'Función	'],
        ['Nº pers.', 'Soc.', 'Clave de organización', 'Un.org.', 'Función'],
        ['Nº pers.', ' Función', 'Ubicación', 'Clv Edo Civ', 'Clv Genero' ],
        ['Nº pers.', 'Soc.', 'Div', 'Clv Org', 'Un.org.', 'Posición', 'Función'],
        ['No. Empleado'],
        ['No Personal', 'Sociedad', 'Clv Org', 'Rel Lab', 'Función']]

def toObject(df, cols):
    for i in range(0, len(cols)):
        df[cols[i]] =  df[cols[i]].astype(object)

        return df
    
df_d_r = toObject(df_d_r, cols[0])
df_h_r = toObject(df_h_r, cols[1])
df_d_a = toObject(df_d_a, cols[2])
df_h_a = toObject(df_h_a, cols[3])
df_d_p = toObject(df_d_p, cols[4])
df_h_p = toObject(df_h_p, cols[5])
```

Definición de función para limpieza e imputación

```python
# Definición de función para limpieza e imputación
def preprocessing(df):
    # Manejo de valores repetidos
    df = df.drop_duplicates()

    # Eliminar filas con valores categóricos faltantes
    df = df.dropna(subset=df.select_dtypes(include='object').columns)

    # Imputación de valores numéricos utilizando un modelo de regresión lineal
    imputer = SimpleImputer(strategy='mean')
    model = LinearRegression()

    for col in df.select_dtypes(include='number').columns:
        col_missing = df[col].isnull()
        if col_missing.sum() > 0:
            X_train = df.loc[~col_missing].drop(col, axis=1)
            y_train = df.loc[~col_missing, col]
            X_test = df.loc[col_missing].drop(col, axis=1)

            model.fit(X_train, y_train)
            df.loc[col_missing, col] = model.predict(X_test)

    return df
```

```python
# Llamado de función
dataframes = [df_d_r, df_h_r, df_d_a, df_h_a, df_d_p, df_h_p]

df_d_r = preprocessing(df_d_r)
df_h_r = preprocessing(df_h_r)
df_d_a = preprocessing(df_d_a)
df_h_a = preprocessing(df_h_a)
df_d_p = preprocessing(df_d_p)
df_h_p = preprocessing(df_h_p)
```

```python
# Definición de función para manejo de outliers
def iqr(df, cols):

    # Manejo de outliers por el método del rango intercuartílico
    for col in  range(0, len(cols)):
        Q1 = df[cols[col]].quantile(0.25)
        Q3 = df[cols[col]].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Eliminar outliers
        df = df[(df[cols[col]] >= lower_bound) & (df[cols[col]] <= upper_bound)]
    
    return df

df_d_p = iqr(df_d_p, ['Edad Ingreso', 'Edad Jubilación'])
```

```python
# Columna posiciones totales
dataframes = [df_d_r, df_h_r, df_d_a, df_h_a, df_d_p, df_h_p]

def total_positions(df, i):
    # Contar las apariciones de cada id en el segundo DataFrame
    count_h = dataframes[i + 1].groupby('id').size().reset_index(name='posiciones totales')

    # Unir los DataFrames usando merge en la columna 'id'
    dataframes[i] = pd.merge(dataframes[i], count_h, on='id', how='left')

    # Rellenar con 0 las filas donde no hay coincidencia en df2
    dataframes[i]['posiciones totales'] = dataframes[i]['posiciones totales'].fillna(1).astype(int)

    return dataframes[i]

df_d_r = total_positions(df_d_r, 0)
df_d_a = total_positions(df_d_a, 2)
df_d_p = total_positions(df_d_p, 4)
```

```python
# Columna movimientos reales
dataframes = [df_h_r, df_h_a, df_h_p]
cols = [['id', ' Función', ' Un.org.', 'Ubicación', 'Desc APers', 'Desc Div Per'],
        ['id', 'Desc Soc', 'Desc Div', 'Desc SubD Per', 'DivP', 'ÁPers', 'Desc Fun'],
        ['id', 'Desc Soc', 'Div Personal', 'Desc Gpo Pers', 'Función', 'Desc Div Personal']]

def real_positions(df, i):
    df_aux = dataframes[i].copy()

    # Calcular registros únicos basados en columnas de interés
    df_aux = df_aux.drop_duplicates(subset=cols[i])

    # Contar registros únicos por ID
    aux = df_aux['id'].value_counts().reset_index()
    aux.columns = ['id', 'movimientos reales']

    # Combinar la información de conteo con el dataframe demographic
    df = df.merge(aux, on='id', how='left')

    # Rellenar los valores NaN con 0 en la columna 'Movimientos reales'
    df['movimientos reales'] = df['movimientos reales'].fillna(1).astype(int)

    return df

df_d_r = real_positions(df_d_r, 0)
df_d_a = real_positions(df_d_a, 1)
df_d_p = real_positions(df_d_p, 2)
```

```python
# columna dias por posicion
dataframes = [df_h_r, df_h_a, df_h_p]

def days_per_position(df, key):
    # Convertir las columnas 'Desde' y 'Hasta' al formato de fecha
    dataframes[key]['Desde'] = pd.to_datetime(dataframes[key]['Desde'], format='mixed', errors='coerce')
    dataframes[key]['Hasta'] = pd.to_datetime(dataframes[key]['Hasta'], format='mixed', errors='coerce')

    # Eliminar filas con fechas inválidas
    dataframes[key] = dataframes[key].dropna()

    # Calcular la diferencia en días y agregar una nueva columna 'dias por posicion'
    dataframes[key]['dias por posicion'] = (dataframes[key]['Hasta'] - dataframes[key]['Desde']).dt.days

    # Calcular el promedio de la columna 'dias por posicion' por 'id'
    average = dataframes[key].groupby('id')['dias por posicion'].mean().reset_index()

    # Unir los DataFrames por 'id'
    df = pd.merge(df, average, on='id', how='left')

    # Rellenar los valores NaN con 0 en la columna 'Movimientos reales'
    df['dias por posicion'] = df['dias por posicion'].fillna(0).astype(int)

    # Ajustar a un día mínimo por posición
    df.loc[df['dias por posicion'] == 0, 'dias por posicion'] = 1

    return df

df_d_r = days_per_position(df_d_r, 0)
df_d_a = days_per_position(df_d_a, 1)
df_d_p = days_per_position(df_d_p, 2)
```

```python
# Columna liderazgo
# Función para calcular las posiciones antes de alcanzar un puesto de liderazgo
def achieved_leadership(group):

    keywords = ['Ejecutivo', 
            'Director', 
            'Gerente', 
            'Jefe', 
            'Administrador', 
            'Responsable', 
            'Encargado', 
            'Subdirector']

    for index, row in group.iterrows():
        if any(keyword in row['Función'] for keyword in keywords):
            return True

    return False

def get_leadership(df, df_his, column):

    # Copiar dataframes originales
    target = df.copy()
    df_aux = df_his.copy()

    # Cambiar nombre a la columna buscada
    df_aux.rename(columns={column: 'Función'}, inplace=True)

    # Guardar resultado por id
    result = df_aux.groupby('id').apply(achieved_leadership).reset_index(name='liderazgo')

    # Combinar los resultados con el dataframe original usando 'Nº pers' como clave de ordenamiento
    target = target.merge(result, on='id', how='left')

    target['liderazgo'] = target['liderazgo'].fillna(False).astype(bool)

    return target

df_d_r = get_leadership(df_d_r, df_h_r, 'Desc Función')
df_d_a = get_leadership(df_d_a, df_h_a, 'Desc Fun')
df_d_p = get_leadership(df_d_p, df_h_p, 'Desc Función')
```

```python
# Función para calcular los días hasta el primer puesto de liderazgo
def days_to_first_leadership(group):

    keywords = ['Ejecutivo', 
            'Director', 
            'Gerente', 
            'Jefe', 
            'Administrador', 
            'Responsable', 
            'Encargado', 
            'Subdirector']

    days = 0
    
    group = group.sort_values(by='Desde')

    for index, row in group.iterrows():
        if any(keyword in row['Función'] for keyword in keywords):
            return days
        days += (row['Hasta'] - row['Desde']).days

    return -1

def get_days_to_leadership(df, df_his, column):

    # Copiar dataframes originales
    target = df.copy()
    df_aux = df_his.copy()

    # Cambiar nombre a la columna buscada
    df_aux.rename(columns={column: 'Función'}, inplace=True)

    # Guardar resultado por id
    result = df_aux.groupby('id').apply(days_to_first_leadership).reset_index(name='dias hasta liderazgo')

    # Combinar los resultados con el dataframe original usando 'Nº pers' como clave de ordenamiento
    target = target.merge(result, on='id', how='left')

    target['dias hasta liderazgo'] = target['dias hasta liderazgo'].fillna(0).astype(int)

    return target

df_d_r = get_days_to_leadership(df_d_r, df_h_r, 'Desc Función')
df_d_a = get_days_to_leadership(df_d_a, df_h_a, 'Desc Fun')
df_d_p = get_days_to_leadership(df_d_p, df_h_p, 'Desc Función')
```

```python
# Columna posiciones hasta liderazgo
def positions_before_leadership(group):

    keywords = ['Ejecutivo', 
            'Director', 
            'Gerente', 
            'Jefe', 
            'Administrador', 
            'Responsable', 
            'Encargado', 
            'Subdirector']

    positions = 1

    group = group.sort_values(by='Desde')

    for index, row in group.iterrows():
        if any(keyword in row['Función'] for keyword in keywords):
            return positions
        positions += 1

    return -1

def get_positions_to_leadership(df, df_his, column):

    # Copiar dataframes originales
    target = df.copy()
    df_aux = df_his.copy()

    # Cambiar nombre a la columna buscada
    df_aux.rename(columns={column: 'Función'}, inplace=True)

    # Guardar resultado por id
    result = df_aux.groupby('id').apply(positions_before_leadership).reset_index(name='posiciones hasta liderazgo')

    # Combinar los resultados con el dataframe original usando 'Nº pers' como clave de ordenamiento
    target = target.merge(result, on='id', how='left')

    target['posiciones hasta liderazgo'] = target['posiciones hasta liderazgo'].fillna(0).astype(int)

    return target

df_d_r = get_positions_to_leadership(df_d_r, df_h_r, 'Desc Función')
df_d_a = get_positions_to_leadership(df_d_a, df_h_a, 'Desc Fun')
df_d_p = get_positions_to_leadership(df_d_p, df_h_p, 'Desc Función')
```

```python
# Columna sindicato 
def sindicate(group):

    keywords = ['Planta Sind.',
                'Eventual Sind.',
                'Vía Planta Sind.']

    group = group.sort_values(by='Desde')

    for index, row in group.iterrows():
        if any(keyword in row['Función'] for keyword in keywords):
            return True

    return False

def getSindicate(df, df_his, column):

    # Copiar dataframes originales
    target = df.copy()
    df_aux = df_his.copy()

    # Cambiar nombre a la columna buscada
    df_aux.rename(columns={column: 'Función'}, inplace=True)

    # Guardar resultado por id
    result = df_aux.groupby('id').apply(sindicate).reset_index(name='sindicato')

    # Combinar los resultados con el dataframe original usando 'Nº pers' como clave de ordenamiento
    target = target.merge(result, on='id', how='left')

    target['sindicato'] = target['sindicato'].fillna(False).astype(bool)

    return target

df_d_r = getSindicate(df_d_r, df_h_r, 'Desc Gpo Personal')
df_d_a = getSindicate(df_d_a, df_h_a, 'Desc Gpo Pers')
df_d_p = getSindicate(df_d_p, df_h_p, 'Desc Gpo Pers')
```
